```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# create training dataset
norm_samples = 1000
anorm_samples = round(norm_samples/100) if round(norm_samples/100) else 5
    
# kilometers from last purchase (kmflp)
mu = 0
sigma = 10
kmflp_norm = np.random.normal(mu, sigma, norm_samples)
kmflp_norm = np.absolute(kmflp_norm)
kmflp_anorm = np.random.uniform(3, 500, anorm_samples)

plt.hist(kmflp_norm, 30)
plt.hist(kmflp_anorm, 30)
plt.show()

# amount (dollars)
mu = 150
sigma = 300
amnt_norm = np.random.normal(mu, sigma, norm_samples)
amnt_norm = np.absolute(amnt_norm)
amnt_anorm = np.random.uniform(30, 5000, anorm_samples)

plt.hist(amnt_norm, 30)
plt.hist(amnt_anorm, 30)
plt.show()

# concatinate to form training data set
kmflp = np.concatenate((kmflp_norm, kmflp_anorm))
amnt = np.concatenate((amnt_norm, amnt_anorm))
data = np.concatenate((kmflp.reshape(-1,1), amnt.reshape(-1,1)), axis=1)

plt.plot(data[:, 0], data[:, 1], 'x')
plt.show()
```

```python
# create test/validation dataset
test_norms = 50
test_anorms = round(test_norms/5) if round(test_norms/5) else 5
y = np.concatenate((np.zeros(test_norms), np.ones(test_anorms)))
    
# kilometers from last purchase (kmflp)
mu = 0
sigma = 10
kmflp_test_norms = np.random.normal(mu, sigma, test_norms)
kmflp_test_norms = np.absolute(kmflp_test_norms)
kmflp_test_anorms = np.random.uniform(3, 500, test_anorms)

plt.hist(kmflp_test_norms, 30)
plt.hist(kmflp_test_anorms, 30)
plt.show()

# amount (dollars)
mu = 150
sigma = 300
amnt_test_norm = np.random.normal(mu, sigma, test_norms)
amnt_test_norm = np.absolute(amnt_test_norm)
amnt_test_anorm = np.random.uniform(30, 5000, test_anorms)

plt.hist(amnt_test_norm, 30)
plt.hist(amnt_test_anorm, 30)
plt.show()

# concatinate to form training data set
kmflp_test = np.concatenate((kmflp_test_norms, kmflp_test_anorms))
amnt_test = np.concatenate((amnt_test_norm, amnt_test_anorm))
data_test = np.concatenate((kmflp_test.reshape(-1,1), amnt_test.reshape(-1,1), y.reshape(-1,1)), axis=1)

plt.plot(kmflp_test_norms, amnt_test_norm, 'x')
plt.plot(kmflp_test_anorms, amnt_test_anorm, 'rx')
plt.show()
```
