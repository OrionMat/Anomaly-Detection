```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from numpy.random import RandomState
```

```python
X = np.loadtxt('transactions.csv', delimiter=',')
labelled = np.loadtxt('transactions_test.csv', delimiter=',')

seed = RandomState()
Xval, Xtest, yval, ytest = train_test_split(labelled[:, [0, 1]], labelled[:, 2], train_size=0.7, random_state=seed)
print('X shape:', X.shape)
print('Xval shape:', Xval.shape)
print('Xtest shape:', Xtest.shape)
```

```python
plt.hist(X[:, 0], 150)
plt.show()
plt.hist(X[:, 1], 150)
plt.show()
```

```python
# transforming data to exhibit more normal characteristics
Xtrain = np.power(X, 1/3)
plt.hist(Xtrain[:, 0], 150)
plt.show()
plt.hist(Xtrain[:, 1], 150)
plt.show()
```

```python
# fit the normal PDF to the data (single variate)
mu1, std1 = norm.fit(Xtrain[:, 0])
mu2, std2 = norm.fit(Xtrain[:, 1])

print(mu1, std1)
print(mu2, std2)
```

```python
# plot PDFs
plt.hist(Xtrain[:, 0], bins=150, density=True)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mu1, std1)
plt.plot(x, y)
plt.show()

plt.hist(Xtrain[:, 1], bins=150, density=True)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, mu2, std2)
plt.plot(x, y)
plt.show()
```

```python
# plot validation and test set
plt.plot(Xval[:, 0], Xval[:, 1], 'x')
plt.show()
plt.plot(Xtest[:, 0], Xtest[:, 1], 'x')
plt.show()
```

```python
# fit the anomaly threshold (epsilon) using validation set
Xval = np.power(Xval, 1/3)

p1 = norm.pdf(Xval[:, 0], mu1, std1)
p2 = norm.pdf(Xval[:, 1], mu2, std2)
p = np.multiply(p1, p2)

pmin, pmax = np.min(p), np.max(p)
step = (pmax - pmin)/1000

maxF1 = 0
epBest = (pmin + pmax)/2
for epsilon in np.arange(pmin, pmax, step):
    predictions = p < epsilon
    f1Score = f1_score(preds, yval)
    if f1Score > maxF1:
        maxF1 = f1Score
        epBest = epsilon
        
print('threshold: ', epBest)
print('f1 score: ', maxF1)
```

```python
# plot validation and test set
plt.plot(Xval[:, 0], Xval[:, 1], 'x')
plt.show()
plt.plot(Xtest[:, 0], Xtest[:, 1], 'x')
plt.show()
```

```python
x1 = np.arange(0, 500, 2)
x2 = np.arange(0, 5000, 20)
X1, X2 = np.meshgrid(x1, x2)

X1trans = np.power(X1, 1/3)
X2trans = np.power(X2, 1/3)

p1mesh = norm.pdf(X1trans, mu1, std1)
p2mesh = norm.pdf(X2trans, mu2, std2)
pmesh = np.multiply(p1mesh, p2mesh)

plt.contour(X1trans, X2trans, pmesh, 10)
plt.plot(Xval[:, 0], Xval[:, 1], 'x')
plt.show()

Xtest = np.power(Xtest, 1/3)
plt.contour(X1trans, X2trans, pmesh, 10)
plt.plot(Xtest[:, 0], Xtest[:, 1], 'x')
plt.show()
```
